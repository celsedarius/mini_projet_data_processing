<h2> <font color='green'> 3. PRETRAITREMENT </font></h2>

### <font color='blue'> Import </font>

```python
import pandas as pd

data= pd.read_csv("BF/BF.txt", sep="\t", header=None) 
##on met le nom original des colonnes
data.columns=["geonameid", "name" ,"asciiname" ,"alternatenames","latitude" ,"longitude" ,"feature class","feature code" ,"country code","cc2","admin1 code" ,"admin2 code" ,"admin3 code","admin4 code","population","elevation","dem" ,"timezone","modification date"]
data
```

###  <font color='blue'> Filtrage de donnees et reindexage des colonnes selectionnees </font>

```python
#on selectionne les colonnes qui nous interressent
burkina= data[["geonameid","name","latitude","longitude"]]

# on renomme les colonnes
burkina.columns=["ID", 'location_name', 'lat', 'long']
burkina
```

### <font color='blue'>  Sauvegarde des donnees dans le dossier output </font>

```python
#on enregistre la selectionne dans le dossier output
burkina.to_csv('output/burkina_location.csv', index=False) 
```

<h2> <font color='green'> 4. Opérations sur le fichier CSV burkina_location.csv </font></h2>

### <font color='blue'> Extraction des donnnes contenant "gounghin" </font>

```python
#on met la colonne en minuscule afin de verifier la presence de gounghin
burkina_lower=burkina['location_name'].str.lower()
gounghin_validator=burkina_lower.str.contains('gounghin')

#on affiche les donnees qui contiennent "gounghin"
burkina[gounghin_validator]
```

### <font color='blue'> Enregistrement de la selection </font>

```python
burkina_location =pd.read_csv("output/burkina_location.csv") 
#enregistrement de la selection dans le dossier output
burkina_location[gounghin_validator].to_csv('output/gounghin.csv', index=False) 
```

### <font color='blue'> Selection des localites dont les noms commencent entre A et P <font>

```python
#on recupere la liste des localites dont le nom commence par ordre alphabetique entre A et P
lis_A_to_P=burkina_location[burkina_location['location_name'].str.startswith(("A","B","C","D","E","F","G","H","I","J","K","L","M","N","P"))]
```

```python
## ordonner par ordre alphabetique 
list_A_to_P= lis_A_to_P.sort_values('location_name')
list_A_to_P
```

### <font color='blue'>  Presentation de la localite avec la latitude minimale </font>

```python
#on recupere l'index de la colonne ayant la lattitude minimal que l'on place dans un iloc
min_lat= burkina_location.iloc[burkina_location["lat"].idxmin()]
min_lat
```

### <font color='blue'> Presentation de la localite avec la longitude minimale </font>

```python
min_long= burkina_location.iloc[burkina_location["long"].idxmin()]
min_long
```

### <font color='blue'>  Listes des localites dont les coordonnnees respectent les conditions </font>

```python
lat_validator= burkina_location['lat']>=11
long_validator= burkina_location['long']<=0.5

list_coordonnee= burkina_location[lat_validator & long_validator]
list_coordonnee
```

<h2> <font color='green'>  5. Sorties Excel </font></h2>

```python
## ecriture des 2 feuilles dans le fichier excel

with pd.ExcelWriter('output/mini_projet.xlsx') as writer:  
    burkina_location[gounghin_validator].to_excel(writer, sheet_name='gounghin', index=False)
    list_A_to_P.to_excel(writer, sheet_name='A_to_P', index=False) 
```
